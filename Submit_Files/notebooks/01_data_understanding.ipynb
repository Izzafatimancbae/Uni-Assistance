## 4. Target Variable Definition

We define a binary target variable:
- **At_Risk = 1** → `final_result` is 'Fail' or 'Withdrawn'
- **At_Risk = 0** → `final_result` is 'Pass' or 'Distinction'

This is appropriate because:
- Withdrawn students did not complete the course
- Failing students did not meet minimum requirements
- Both groups require academic intervention

```python
student_info['At_Risk'] = student_info['final_result'].map(
    {'Fail': 1, 'Withdrawn': 1, 'Pass': 0, 'Distinction': 0}
)
print('At-Risk distribution:')
print(student_info['At_Risk'].value_counts())
print(f'\nAt-Risk rate: {student_info["At_Risk"].mean()*100:.1f}%')
```

## 5. Missing Value Summary

```python
dfs = {'studentInfo': student_info, 'studentAssessment': student_assessment,
       'assessments': assessments, 'studentVle': student_vle,
       'studentRegistration': student_registration}

for name, df in dfs.items():
    missing = df.isnull().sum().sum()
    print(f'{name}: {missing} missing values')
```
